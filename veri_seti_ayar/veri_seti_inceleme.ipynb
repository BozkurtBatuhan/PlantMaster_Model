```python
import os
```

```python
data_dir = 'plant_village_dataset/train_val_test/test/'
```

```python
class_counts = {}

for class_name in os.listdir(data_dir):
    class_path = os.path.join(data_dir, class_name)
    
    if os.path.isdir(class_path):
        num_images = len(os.listdir(class_path))
        class_counts[class_name] = num_images
```

```python
sorted_counts = dict(sorted(class_counts.items(), key=lambda x: x[1], reverse=True))
```

```python
for cls, count in sorted_counts.items():
    print(f'{cls}: {count}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.bar(sorted_counts.keys(), sorted_counts.values())
plt.xticks(rotation=90)
plt.ylabel("Image Count")
plt.title("Class Distribution in PlantVillage Dataset")
plt.tight_layout()
plt.show()
```
